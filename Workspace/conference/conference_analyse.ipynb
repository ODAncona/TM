```python
tresh_score = 8
tresh_deadline = pd.Timestamp('2025.05.01')


def load_data(path:str, tresh_score:int, tresh_deadline:str) -> pd.DataFrame:
    data = pd.read_csv(path)
    data['Deadline'] = pd.to_datetime(data['Deadline'])
    df = data[(data['Score'] >= tresh_score) & (data['Deadline'] >= tresh_deadline)]
    return df.sort_values(by='Score', ascending=False)
     
```

```python
path_computing = "/home/olivier/projet/cfp_crawler/src/cfp_crawler/result_computing.csv"
path_distributed = "/home/olivier/projet/cfp_crawler/src/cfp_crawler/result_distributed.csv"
path_result = "result.csv"

df1 = load_data(path_computing, tresh_score, tresh_deadline)
df2 = load_data(path_distributed, tresh_score, tresh_deadline)
df3 = load_data(path_result, tresh_score, tresh_deadline)
df = pd.concat([df1, df2, df3])
```

```python
df = df.drop_duplicates(subset=['Title'])
```
